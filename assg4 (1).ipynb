```python
import requests
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"

try:
    response = requests.get(url)
    response.raise_for_status()  # Raise an HTTPError for bad requests

    soup = BeautifulSoup(response.text, 'html.parser')

    # Extract the details you need using BeautifulSoup
    # Replace the following with the actual HTML structure from the Wikipedia page
    rank = soup.find('span', class_='rank').text
    name = soup.find('span', class_='name').text
    artist = soup.find('span', class_='artist').text
    upload_date = soup.find('span', class_='upload-date').text

    print(f"Rank: {rank}\nName: {name}\nArtist: {artist}\nUpload Date: {upload_date}")

except requests.exceptions.HTTPError as http_err:
    print(f"HTTP error occurred: {http_err}")

except Exception as err:
    print(f"An error occurred: {err}")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

url = "https://www.bcci.tv/"

# Create a new instance of the Firefox driver (you can use other drivers like Chrome or Edge)
driver = webdriver.Firefox()

try:
    # Open the BCCI website
    driver.get(url)

    # Find and click on the "International" link to navigate to the international fixtures page
    international_link = driver.find_element(By.XPATH, "//a[text()='International']")
    international_link.click()

    # Wait for the international fixtures page to load
    WebDriverWait(driver, 10).until(EC.title_contains("Fixtures"))

    # Extract details from the page
    series = driver.find_element(By.XPATH, "//div[@class='fixture__format-strip']")
    place = driver.find_element(By.XPATH, "//p[@class='fixture__additional-info']/span")
    date = driver.find_element(By.XPATH, "//span[@class='fixture__datetime tablet-only']/strong")
    time = driver.find_element(By.XPATH, "//span[@class='fixture__time']/strong")

    # Print the extracted details
    print(f"Series: {series.text}")
    print(f"Place: {place.text}")
    print(f"Date: {date.text}")
    print(f"Time: {time.text}")

except Exception as e:
    print(f"An error occurred: {e}")

finally:
    # Close the browser window
    driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

url = "http://statisticstimes.com/"

# Create a new instance of the Firefox driver (you can use other drivers like Chrome or Edge)
driver = webdriver.Firefox()

try:
    # Open the Statistics Times website
    driver.get(url)

    # Find and click on the "Economy" link to navigate to the economy page
    economy_link = driver.find_element(By.XPATH, "//div[@class='navbar']/div[2]/div/a[3]")
    economy_link.click()

    # Find and click on the "GDP of Indian states" link to navigate to the state-wise GDP page
    gdp_link = driver.find_element(By.XPATH, "//ul[@style='float:left;']/li[1]/a")
    gdp_link.click()

    # Wait for the state-wise GDP page to load
    WebDriverWait(driver, 10).until(EC.title_contains("GDP of Indian states"))

    # Extract details from the page
    ranks = driver.find_elements(By.XPATH, "//td[@class='data1']")
    states = driver.find_elements(By.XPATH, "//td[@class='name']")
    gsdp_1819 = driver.find_elements(By.XPATH, "//td[@class='data sorting_1']")
    gsdp_1920 = driver.find_elements(By.XPATH, "//td[@class='data']")
    share_1819 = driver.find_elements(By.XPATH, "//td[@class='data']")[1::4]
    gdp_billion = driver.find_elements(By.XPATH, "//td[@class='data']")[3::4]

    # Print the extracted details
    for i in range(len(states)):
        print(f"Rank: {ranks[i].text}")
        print(f"State: {states[i].text}")
        print(f"GSDP(18-19) - at current prices: {gsdp_1819[i].text}")
        print(f"GSDP(19-20) - at current prices: {gsdp_1920[i].text}")
        print(f"Share(18-19): {share_1819[i].text}")
        print(f"GDP($ billion): {gdp_billion[i].text}")
        print("\n")

except Exception as e:
    print(f"An error occurred: {e}")

finally:
    # Close the browser window
    driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

url = "https://github.com/"

# Create a new instance of the Firefox driver (you can use other drivers like Chrome or Edge)
driver = webdriver.Firefox()

try:
    # Open the GitHub website
    driver.get(url)

    # Find and click on the "Explore" menu
    explore_menu = driver.find_element(By.XPATH, "//summary[contains(text(),'Explore')]")
    explore_menu.click()

    # Find and click on the "Trending" option under Explore menu
    trending_option = driver.find_element(By.XPATH, "//a[text()='Trending']")
    trending_option.click()

    # Wait for the trending repositories page to load
    WebDriverWait(driver, 10).until(EC.title_contains("Trending"))

    # Extract details from the page
    repository_titles = driver.find_elements(By.XPATH, "//h1[@class='h3 lh-condensed']/a")
    repository_descriptions = driver.find_elements(By.XPATH, "//p[@class='col-9 color-text-secondary my-1 pr-4']")
    contributors_counts = driver.find_elements(By.XPATH, "//a[contains(@href, '/network/members')]/following-sibling::span")
    languages_used = driver.find_elements(By.XPATH, "//span[@itemprop='programmingLanguage']")

    # Print the extracted details
    for i in range(len(repository_titles)):
        print(f"Repository Title: {repository_titles[i].text}")
        print(f"Repository Description: {repository_descriptions[i].text}")
        print(f"Contributors Count: {contributors_counts[i].text}")
        print(f"Language Used: {languages_used[i].text}")
        print("\n")

except Exception as e:
    print(f"An error occurred: {e}")

finally:
    # Close the browser window
    driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

url = "https://www.billboard.com/"

# Create a new instance of the Firefox driver (you can use other drivers like Chrome or Edge)
driver = webdriver.Firefox()

try:
    # Open the Billboard website
    driver.get(url)

    # Find and click on the "Charts" option
    charts_option = driver.find_element(By.XPATH, "//a[contains(text(),'Charts')]")
    charts_option.click()

    # Find and click on the "Hot 100" link under Charts
    hot_100_link = driver.find_element(By.XPATH, "//a[text()='Hot 100']")
    hot_100_link.click()

    # Wait for the Hot 100 page to load
    WebDriverWait(driver, 10).until(EC.title_contains("Hot 100"))

    # Extract details from the page
    song_names = driver.find_elements(By.XPATH, "//span[@class='chart-element__information__song text--truncate color--primary']")
    artist_names = driver.find_elements(By.XPATH, "//span[@class='chart-element__information__artist text--truncate color--secondary']")
    last_week_ranks = driver.find_elements(By.XPATH, "//span[@class='chart-element__meta text--center color--secondary text--last']")
    peak_ranks = driver.find_elements(By.XPATH, "//span[@class='chart-element__meta text--center color--secondary text--peak']")
    weeks_on_board = driver.find_elements(By.XPATH, "//span[@class='chart-element__meta text--center color--secondary text--week']")

    # Print the extracted details for the top 100 songs
    for i in range(len(song_names)):
        print(f"Song Name: {song_names[i].text}")
        print(f"Artist Name: {artist_names[i].text}")
        print(f"Last Week Rank: {last_week_ranks[i].text}")
        print(f"Peak Rank: {peak_ranks[i].text}")
        print(f"Weeks on Board: {weeks_on_board[i].text}")
        print("\n")

except Exception as e:
    print(f"An error occurred: {e}")

finally:
    # Close the browser window
    driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

url = "https://www.imdb.com/list/ls095964455/"

# Create a new instance of the Firefox driver (you can use other drivers like Chrome or Edge)
driver = webdriver.Firefox()

try:
    # Open the IMDb URL
    driver.get(url)

    # Wait for the page to load
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "lister-list")))

    # Extract details from the page
    names = driver.find_elements(By.XPATH, "//h3[@class='lister-item-header']/a")
    year_spans = driver.find_elements(By.XPATH, "//h3[@class='lister-item-header']/span[2]")
    genres = driver.find_elements(By.XPATH, "//p[@class='text-muted text-small']/span[@class='genre']")
    run_times = driver.find_elements(By.XPATH, "//p[@class='text-muted text-small']/span[@class='runtime']")
    ratings = driver.find_elements(By.XPATH, "//div[@class='ipl-rating-widget']/div[@class='ipl-rating-star small']/span[@class='ipl-rating-star__rating']")
    votes = driver.find_elements(By.XPATH, "//p[@class='text-muted text-small']/span[@name='nv']")

    # Print the extracted details for the most-watched TV series
    for i in range(len(names)):
        print(f"Name: {names[i].text}")
        print(f"Year Span: {year_spans[i].text}")
        print(f"Genre: {genres[i].text.strip()}")
        print(f"Run Time: {run_times[i].text}")
        print(f"Ratings: {ratings[i].text}")
        print(f"Votes: {votes[i].text}")
        print("\n")

except Exception as e:
    print(f"An error occurred: {e}")

finally:
    # Close the browser window
    driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

url = "https://archive.ics.uci.edu/"

# Create a new instance of the Firefox driver (you can use other drivers like Chrome or Edge)
driver = webdriver.Firefox()

try:
    # Open the UCI Machine Learning Repository URL
    driver.get(url)

    # Find and click on the "View All Data Sets" link to navigate to the dataset page
    view_all_link = driver.find_element(By.XPATH, "//a[contains(text(),'View All Data Sets')]")
    view_all_link.click()

    # Wait for the dataset page to load
    WebDriverWait(driver, 10).until(EC.title_contains("View ALL Data Sets"))

    # Extract details from the page
    dataset_names = driver.find_elements(By.XPATH, "//table[@cellpadding='3']/tbody/tr/td[2]/p/b/a")
    data_types = driver.find_elements(By.XPATH, "//table[@cellpadding='3']/tbody/tr/td[3]/p")
    tasks = driver.find_elements(By.XPATH, "//table[@cellpadding='3']/tbody/tr/td[4]/p")
    attribute_types = driver.find_elements(By.XPATH, "//table[@cellpadding='3']/tbody/tr/td[5]/p")
    no_of_instances = driver.find_elements(By.XPATH, "//table[@cellpadding='3']/tbody/tr/td[6]/p")
    no_of_attributes = driver.find_elements(By.XPATH, "//table[@cellpadding='3']/tbody/tr/td[7]/p")
    years = driver.find_elements(By.XPATH, "//table[@cellpadding='3']/tbody/tr/td[8]/p")

    # Print the extracted details for the datasets
    for i in range(len(dataset_names)):
        print(f"Dataset Name: {dataset_names[i].text}")
        print(f"Data Type: {data_types[i].text}")
        print(f"Task: {tasks[i].text}")
        print(f"Attribute Type: {attribute_types[i].text}")
        print(f"No of Instances: {no_of_instances[i].text}")
        print(f"No of Attributes: {no_of_attributes[i].text}")
        print(f"Year: {years[i].text}")
        print("\n")

except Exception as e:
    print(f"An error occurred: {e}")

finally:
    # Close the browser window
    driver.quit()
```
